```python
import matplotlib.pyplot as plt
import numpy as np
from pydicom import dcmread
```

```python


_cw = r"E:\CMC\pyprojects\radio_therapy\dose-3d\dataset\SVMAT\RECON_SIB COMP_26.7.25\CW\cw.dcm"
_cw1 = r'E:\CMC\pyprojects\radio_therapy\dose-3d\dataset\SVMAT\RECON_SIB COMP_26.7.25\CW1\cw1.dcm'
```

```python
cw = dcmread(_cw).pixel_array
cw1 = dcmread(_cw1).pixel_array
```

```python
rec_image = cw + cw1

# rec_image = np.transpose(rec_image, (2, 1, 0))  # Transpose to match the expected orientation
Nimage = 100
Z_c=int(Nimage/2.0)
X_c=int(Nimage/2.0)
Y_c=int(Nimage/2.0)
figure, axis=plt.subplots(1, 3,figsize=(12,12))
axis[0].imshow(rec_image[50,:,:].T, cmap='viridis')
axis[0].set_title('cross section at X=0')
axis[0].axis('off')
axis[1].imshow(rec_image[:,Y_c,:].T, cmap='viridis')
axis[1].set_title('cross section at Y=0')
axis[1].axis('off')
axis[2].imshow(rec_image[:,:,50].T, cmap='viridis')
axis[2].set_title('cross section at Z=0')
axis[2].axis('off')
plt.show()
```
